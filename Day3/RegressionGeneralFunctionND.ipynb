Expectation: A working solution where the learned function using Linear Regression is close to the training data points.

```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

pi = math.pi

fig = plt.figure()
ax = fig.gca(projection='3d')
x1 = np.linspace(-1,1,100)
x2 = np.linspace(-1,1,100)
[X1,X2] = np.meshgrid(x1,x2)

Y = np.cos(2*pi*X1)*np.sin(2*pi*1*X2)+0.1*np.random.randn(X1.shape[0],X1.shape[1])

surf = ax.plot_surface(X1, X2, Y, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python
tX = []
num_basis = 7
for i in range(num_basis+1):
  for j in range(num_basis+1):
    X.append(X1.reshape(1,-1)**i * X2.reshape(1,-1)**j)

X = np.array(X).squeeze().T


w = np.linalg.solve(X.T@X, X.T@Y.reshape(-1,1))

yhat = X@w
yhat = yhat.reshape(100,100)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X1, X2, yhat, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python
M = 7
X = []
for i in range(M+1):
  for j in range(M+1):
    X.append(X1.reshape(-1,)**i * X2.reshape(-1,)**j)

X = np.array(X).T
print(X.shape)

theta = np.linalg.solve(X.T@X, X.T@Y.reshape(-1,1))

Yhat = X@theta

Yhat = Yhat.reshape(Y.shape)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X1, X2, Yhat, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
