```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plotClass(X,y,p):
    for i in range(y.shape[1]):
        if y[0,i] == 0:
            plt.plot(X[0,i], X[1,i],'r'+p)
        else:
            plt.plot(X[0,i], X[1,i],'b'+p)            
```

```python
num_data=100 # data points per class

x1=np.random.randn(2,num_data)+4
x0=np.random.randn(2,num_data)

y1=np.ones((1,num_data))
y0=np.zeros((1,num_data))

X=np.concatenate((x1,x0),axis=1)
y=np.concatenate((y1,y0), axis=1)
plotClass(X,y,'o')
print(X.shape)
print(y.shape)
```

```python
X = X.T
y = y.T
```

```python
col = np.ones((X.shape[0], 1))
A = np.concatenate((col, X), axis =1)
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
def loss(yhat, y):
  return np.sum(-y*np.log(yhat) - (1-y)*np.log(1-yhat), axis =0)
```

```python
# Z = X@theta

# w is theta
w = np.random.randn(A.shape[1],1)

lr = 0.001
ls = []

for i in range(1000):
  # X is represented by A 
  Z = A@w
  yhat = sigmoid(Z)
  l = loss(yhat, y)
  ls.append(l)
  #dw represent dL/dw
  # X is represented in the code by A
  dw = A.T @ (yhat - y)
    
  # plotting decison boundary
  if i%100 == 0:
    plt.figure()
    x1 = np.linspace(-2, 5 , 3)
    x2 = (-w[0] - w[1]*x1)/ w[2]
    plotClass(X.T,y.T,'o')
    plt.plot(x1,x2)


  w = w - lr*dw

plt.figure
plt.plot(ls)
```
