```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
N = 21
# generate X
x = np.linspace(0.0, 20, N)
X0 = x.reshape(N, 1)
X = np.column_stack((np.ones(N), X0))
# generate random y
w = np.array([-1.5, 1 / 9.0])
y = w[0] * x + w[1] * np.square(x)
y = y + np.random.normal(0, 1, N) * 2
# solve
theta = la.solve(X.T @ X, X.T @ y)
y_hat = X @ theta
plt.plot(X[:, 1], y, "o")
plt.plot(X[:, 1], y_hat, "-")
plt.show()
```

```python
#Now let's try with adding x^2
N = 21
# generate X
x = np.linspace(0.0, 20, N)
X0 = x.reshape(N, 1)
X1 = X0**2 # adding x^2
X = np.column_stack((np.ones(N), X0, X1))
# generate random y
w = np.array([-1.5, 1 / 9.0])
y = w[0] * x + w[1] * np.square(x)
y = y + np.random.normal(0, 1, N) * 2
# solve
theta = la.solve(X.T @ X, X.T @ y)
y_hat = X @ theta
plt.plot(X[:, 1], y, "o")
plt.plot(X[:, 1], y_hat, "-")
plt.show()
```
